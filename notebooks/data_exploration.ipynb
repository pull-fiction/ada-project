```python
import pandas as pd

import sys
sys.path.append("..")
```

```python
ucdp_path = '../data/ucdp-prio/'
ucdp_file_term = 'ucdp-prio-acd-191.csv'

wars_df = pd.read_csv(ucdp_path + ucdp_file_term)
```

```python
simple_wars_df = wars_df[[
    'location',
    'side_a',
    'side_a_2nd',
    'side_b',
    'side_b_2nd',
    'start_date',
    'start_date2',
    'ep_end_date',
    'region',
    'intensity_level',
    'type_of_conflict'
]]
```

```python
wars = simple_wars_df[simple_wars_df['intensity_level'] == 2]

wars[wars['location'] == 'Afghanistan']
```

```python
wars.loc[:, 'start_date'] = pd.to_datetime(wars['start_date'])
wars.loc[:, 'start_date2'] = pd.to_datetime(wars['start_date2'])
wars.loc[:, 'ep_end_date'] = pd.to_datetime(wars['ep_end_date'])
```

```python
wars.loc[:, 'start_year'] = wars['start_date'].map(lambda date: date.year)
wars.loc[:, 'ep_end_year'] = wars['ep_end_date'].map(lambda date: date.year)
wars.loc[:, 'start_year2'] = wars['start_date2'].map(lambda date: date.year)
```

```python
wars = wars.drop(['start_date', 'start_date2', 'ep_end_date'], axis=1)
```

```python
wars.groupby(['location', 'start_year', 'start_year2']).count().head()
```
